```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df = pd.read_csv('df_combined.csv')
```

```python
df.head()
```

```python
df['Stage_Progress'].value_counts()
```

There 200 (66.7%) patients with False stage progress while 100 (33.3%) patients with True stage progress.

### Race

```python
df['race'].value_counts()
```

```python
df[df['race'] == 'Unknown']['id'].nunique()
```

```python
# number of Unknown race when stage progress is True
df[(df['race'] == 'Unknown') & (df['Stage_Progress'] == True)]['id'].nunique()
```

According to the National Kidney Foundation, race is one of the important risk factors for kidney disease.

For our datasets, there is 26 (8.67%) of __Unknown__ race data, I decide to keep them and put them under others because the rest of the data is useful for the prediction.

```python
df[df['Stage_Progress'] == True]['race'].value_counts()
```

```python
df[df['Stage_Progress'] == False]['race'].value_counts()
```

```python
df['race'].value_counts()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
ax = sns.countplot(x="race", hue="Stage_Progress", data=df)
plt.show()
```

Finding:

- Out of 226 White - 76 (34%) have True Stage Progress and 150 (66%) have False.
- Out of 24 Black - 6 (25%) have True Stage Progress and 18 (75%) have False.
- Out of 17 Asian - 5 (29%) have True Stage Progress and 12 (71%) have False.
- Out of 7 Hispanic - 3 (43%) have True Stage Progress and 4 (57%) have False.

### Gender

```python
df['gender'].value_counts()
```

```python
# percentage of male with True Stage Progress
df[(df['Stage_Progress'] == True) & (df['gender'] == 'Male')]['gender'].value_counts() / df[df['gender'] == 'Male']['gender'].value_counts() * 100
```

```python
# percentage of female with True Stage Progress
df[(df['Stage_Progress'] == True) & (df['gender'] == 'Female')]['gender'].value_counts() / df[df['gender'] == 'Female']['gender'].value_counts() * 100
```

From the information above, we can infer that female has a lesser percentage to progress with CKD in a future time point compare to male.

```python
df[df['Stage_Progress'] == True]['gender'].value_counts()
```

From the information above, we can infer that male has a higher rate of True stage progress compare to the female.

```python
f, ax = plt.subplots(figsize=(8, 6))
ax = sns.countplot(x="gender", hue="Stage_Progress", data=df)
plt.show()
```

Finding:

- Out of 176 females - 44 (25%) have True Stage Progress and 132 (75%) have False.

- Out of 124 males - 56 (45%) have True Stage Progress and 68 (55%) have False.

### Age

```python
print('Age aggregation, Stage Progress = True, Gender = Male\n')
print(df[(df['Stage_Progress'] == True) & (df['gender']=='Male')]['age'].describe())
print('-'*88)
print('Age aggregation, Stage Progress = False, Gender = Male\n')
print(df[(df['Stage_Progress'] == False) & (df['gender']=='Male')]['age'].describe())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13,5))

sns.histplot(data=df[df['Stage_Progress'] == True], x='age', bins=10, kde=True, ax=ax1)
ax1.set_title('Histogram of Age when Stage Progress is True', fontsize=15);

sns.histplot(data=df[df['Stage_Progress'] == False], x='age', bins=10, kde=True, ax=ax2)
ax2.set_title('Histogram of Age when Stage Progress is False', fontsize=15)

plt.tight_layout();
```

Seems like both graphs show almost similar information as both peaks is between age 65 to 75.

It seems that most of the latest reading (Glucose, SBP, LDL and DBP) and eGFR are the top correlation with the stage progress.

```python
# use heatmap to find the correlation
def heatmap(df):
    fix, ax = plt.subplots(figsize=(30,30))

    mask = np.zeros_like(df.corr(), dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    ax = sns.heatmap(df.corr(), mask=mask, ax=ax, cmap='coolwarm',vmin=-1, vmax=1,
                     square=True, linewidth=1,annot=True,annot_kws={'size':'13'},
                     cbar_kws={"shrink": 0.8}, fmt='.1g',center=0)

    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=14)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=14)

    ax.set_title('Correlation Heatmap between Stage Progress and Input Variables', fontsize=22);
```

```python
heatmap(df)
```

Looking at the heat map above, we can observe that mean_hgb with min_hgb/ max_hgb and mean_ldl with min_ldl and max_ldl generally have multicollinearity value = 1, which is 100% multicollinearity.

This is a problem because when we using too many correlated features, it may cause the model to consider too many unnecessary features, leading to the curse of dimensionality. In other words, a high correlation between our variables might reduce the precision of our estimate coefficients, which might then weaken the statistical power of our model. 

I will do the pairwise correlation to identify the multicollinearity.

### Pairwise Correlation

This method is to identify the correlation between two variables whether they carry similar information or not. Having a perfect pairwise correlation between two variables means that one of the variables is redundant. Keeping only the one that has a higher correlation coefficient with the target will help reduce dimensionality without much loss of information.

```python
corr_matrix = df.corr().abs()

sol = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1)
                  .astype(np.bool))
                  .stack()
                  .sort_values(ascending=False))

# Convert to dataframe and reset multi-level index
corr_df = pd.DataFrame(sol.head(50)).reset_index()

# Rename columns
corr_df.columns = 'v1', 'v2', 'pair_corr'
```

```python
def corr_target(row):
    row['v1_y_corr'] = df.corr()['Stage_Progress'][row['v1']]
    row['v2_y_corr'] = df.corr()['Stage_Progress'][row['v2']]
    return row
```

```python
corr_df = corr_df.apply(corr_target, axis=1)
```
